```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=correlation(X_train,0.75)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sd=StandardScaler()
```

```python
X_train=sd.fit_transform(X_train)
X_test=sd.transform(X_test)
```

# Selecting Good Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
import xgboost as xb
```

```python
lr=LogisticRegression()
dtc=DecisionTreeClassifier()
rfc=RandomForestClassifier()
svc=SVC()
knc=KNeighborsClassifier()
gnb= GaussianNB()
abc=AdaBoostClassifier()
gbc=GradientBoostingClassifier()
xgc=xb.XGBClassifier()
```

```python
from sklearn.metrics import accuracy_score
```

```python
import warnings
warnings.filterwarnings('ignore')

algos = [lr, dtc, rfc, svc, knc, gnb, abc, gbc, xgc]
names = ["Logistic Regression" , "Decision Tree Classifier" , "Random Forest Classifier" , "Support Vector Classifier" , "K Neighbors Classifier" , "naive bayes classifier", "Ada Boost Classifier", "Gradient Boost Classifier","XG Boost Classifier" ]

for i,j in zip(algos,names):
  i.fit(X_train,y_train)
  y_pred = i.predict(X_test)
  print(j)
  print("accuracy_Score is {}".format(accuracy_score(y_test,y_pred) * 100))
  print("--" * 40)
```

# Hyper_parameter Turning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best Hyperparameters:", grid_search.best_params_)
```

```python
y_pred = grid_search.predict(X_test)
```
